Одномерный массив -  — это последовательность чисел или элементов, которые располагаются подряд в одной строке памяти компьютера. В Python массивы можно создавать с помощью функции list(). Например:

```python
arr = [1, 2, 3, 4, 5]
print(arr)
print(arr[1])
print(arr[1:3])
print(arr[::2])
```

Двумерный массив — это матрица чисел или элементов(также можно назвать структуру списком внутри списка), которые располагаются подряд в одной строке памяти компьютера. В Python массивы можно создавать с помощью функции list().

```python
arr = [
    [1, 2, 3], 
    [4, 5, 6], 
    [7, 8, 9]
    ]

print(arr)
print(arr[0][1])
print(arr[-1])
print(arr[-1][-1])
print(arr[:]) 
```

Напишите программу, которая ищет среди целых чисел, принадлежащих числовому отрезку [126849; 126871], числа, имеющие ровно 4 различных делителя.
Выведите эти четыре делителя для каждого найденного числа в порядке возрастания.

```python
# Неоптимизированный вариант
for n in range(126849,126871+1):
    divs = [] # чистим список делителей
    
    for d in range(1,n+1):
        if n % d == 0:
            divs = divs + [d] # добавляем делитель в список
            # divs = divs.append(d) # добавляем делитель в список
            # if len(divs) > 4: break
    
    if len(divs) == 4:
        # divs.reverse()
        print(*divs)
```

```python
# 1 способ(оптимизированный)
for n in range(126849, 126871+1):
    divs = [d for d in range(1, n+1) if n % d == 0]
    if len(divs) == 4:
        # divs.reverse()
        print(*divs)
```

```python
# 2 способ(оптимизированный)
import math 
divCounter = 4

for n in range(126849, 126871+1):
    divs = []
    for d in range(1, round(math.sqrt(n))):
        if n % d == 0:
            divs.append(d)
            
            if d != n//d:
                divs.append(n//d)
            
            if len(divs) > divCounter: break
    
    if len(divs) == 4:
        divs.sort()
        # divs.reverse()
        print(*divs)
```

Напишите программу, которая ищет среди целых чисел, принадлежащих числовому отрезку [190201; 190280], числа, имеющие ровно 4 различных **ЧЁТНЫХ** делителя.
Выведите эти четыре делителя для каждого найденного числа в порядке убывания.


```python
# Неоптимизированный вариант
for n in range(190201,190280+1):
    divs = []
    for d in range(1,n+1):
        if n % d == 0 and d % 2 == 0:
            divs.append(d) # добавляем делитель в список
        if len(divs) > 4: break
        
    if len(divs) == 4:
        print(*divs)
```

```python
for n in range(190201,190280+1):
    divs = [d for d in range(1,n+1) if n % d == 0 and d % 2 == 0]
    if len(divs) == 4:
        print(*divs)
```
