```python
def f(n):
    if n == 1:
        return 1
    if n == 2:
        return 3
    if n > 2:
        return f(n-1) *n + f(n-2) * (n-1)
print(f(5))
```

```python
def f(n):
    if n in [1,2,3]:
        return 1
    if n > 3:
        return f(n-3) + f(n-2)
print(f(10))
```

```python
def f(n):
    if n == 1:
        return 0
    if n > 1:
        return f(n-1) + n

def g(n):
    if n == 1:
        return 1
    if n > 1:
        return g(n-1) * n

print(f(5) + g(5))
```
